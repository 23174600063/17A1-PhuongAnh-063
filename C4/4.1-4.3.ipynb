### câu 4.1: tạo chuỗi pandas

```python
import pandas as pd 
import numpy as np 

# danh sách, numpy, từ điển
mylists = list('abcedfghijklmnoqprstuvwxyz')
myarr = np.arange(26)
mydict = dict(zip(mylists, myarr))

# tạo chhuỗi pandas

series_list = pd.Series(mylists)
series_arr = pd.Series(myarr)
series_dict = pd.Series(mydict)

print("Chuỗi pandas từ mục danh sách là:",series_list)
print("Chuỗi pandas từ mục numpy là:",series_arr)
print("Chuỗi pandas từ mục dict là:",series_dict)
```

### câu 4.2: chuyển chuỗi ser thành dataframe với chỉ mục của nó dưới dạng một cột khác trên dataframe

```python
import numpy as np
import pandas as pd

# Input
mylist = list('abcdefghijklmnopqrstuvwxyz')  # Danh sách các chữ cái
myarr = np.arange(26)  # Mảng numpy từ 0 đến 25
mydict = dict(zip(mylist, myarr))  # Từ điển ánh xạ chữ cái và số
ser = pd.Series(mydict)  # Chuỗi (Series) từ từ điển

# Chuyển Series thành DataFrame
df = ser.reset_index()  # Chuyển Series thành DataFrame và đưa chỉ mục thành cột riêng
df.columns = ['Index', 'Value']  # Đặt tên cột mới

# In kết quả
print("Chuỗi ban đầu (Series):\n", ser)
print("\nDataFrame sau khi chuyển đổi:\n", df)
```

### câu 4.3: kết nối ser1 và ser2 để tạo thành df


```python


# Input
ser1 = pd.Series(list('abcedfghijklmnopqrstuvwxyz'))  
ser2 = pd.Series(np.arange(26))  

# Kết nối hai Series thành một DataFrame
df = pd.DataFrame({'Column1': ser1, 'Column2': ser2})  # Tạo DataFrame từ hai Series

# In kết quả
print("Series 1:\n", ser1)
print("\nSeries 2:\n", ser2)
print("\nDataFrame sau khi kết nối:\n", df)
```
